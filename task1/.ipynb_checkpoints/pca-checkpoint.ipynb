```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from tool import *

# Generate a random matrix as input
np.random.seed(0)
X = np.random.rand(100, 10)  # 100 samples, 10 features

# Convert the numpy array to a PyTorch tensor
X_tensor = torch.from_numpy(X)
```

```python
X_tensor.shape
```

```python
X_train, X_test, Y_train, Y_test = get_data("./tool")
X_train = torch.from_numpy(X_train)
```

```python
X_train.size()
```

```python
X_tensor = X_train.view(X_train.size(0), -1)
```

```python
X_tensor.size()
```

```python
# Compute the covariance matrix
X_centered = X_tensor - X_tensor.mean(dim=0)
cov_matrix = torch.matmul(X_centered.t(), X_centered) / (X_tensor.shape[0] - 1)

# Compute the eigenvectors and eigenvalues
eigenvalues, eigenvectors = torch.linalg.eig(cov_matrix)
```

```python
eigenvalues
```

```python
# Sort the eigenvalues in descending order
sorted_indices = torch.argsort(eigenvalues.real, descending=True)
eigenvalues_sorted = eigenvalues.real[sorted_indices]
eigenvectors_sorted = eigenvectors.real[:, sorted_indices]

eigenvalues_sorted
```

```python
eigenvalues_sorted.size()
```

```python
# Select the top k eigenvectors based on explained variance
k = 2  # Number of components to keep
components = eigenvectors_sorted[:, :k]

# Project the data onto the selected components
X_pca = torch.matmul(X_centered, components)

# Print the result
print("Original shape:", X_tensor.shape)
print("PCA shape:", X_pca.shape)
```
